```python
import pandas as pd
import numpy as np
```

```python
gameInput = pd.read_csv("nhl-game-data/game.csv").dropna().sort_values(by=['date_time_GMT'])
goalieInput = pd.read_csv("nhl-game-data/game_goalie_stats.csv")
skaterStatsInput = pd.read_csv("nhl-game-data/game_skater_stats.csv")
teamsStatsInput = pd.read_csv("nhl-game-data/game_teams_stats.csv")
teamInfoInput = pd.read_csv("nhl-game-data/team_info.csv")
```

```python
gameInput['date_time_GMT'] = pd.to_datetime(gameInput['date_time_GMT'], errors='coerce')
```

```python
goalies = goalieInput.loc[:,['game_id','team_id','shots','saves']].groupby(['game_id','team_id']).sum().reset_index()
goalies['percentage'] = (goalies['saves'] / goalies['shots'])*100
```

```python
gameInput = gameInput.sort_values(by=['game_id'])
gameInput
```

```python
games = pd.DataFrame(columns=['game_id', 'season', 'time', 'home_scored', 'home_conceded', 'home_win','home_percentage', 'home_id', 'away_scored', 'away_conceded', 'away_win','away_percentage', 'away_id'])
for index, row in gameInput.iterrows():
    homeS = 0;
    homeC = 0;
    homeW = 0;
    awayS = 0;
    awayC = 0;
    awayW = 0;
    homePerc = 50;
    awayPerc = 50;
    if not (gameInput.loc[((gameInput['home_team_id'] == row['home_team_id']) & (gameInput['date_time_GMT'] < row['date_time_GMT']))].tail(10).empty):
        homeLast = gameInput.loc[((gameInput['home_team_id'] == row['home_team_id']) & (gameInput['date_time_GMT'] < row['date_time_GMT']))].tail(10)
        for index1, game in homeLast.iterrows():
            homeS += game['home_goals']
            homeC += game['away_goals']
            if(game['home_goals'] > game['away_goals']):
                homeW += 1
            if not goalies.loc[(goalies['game_id'] == row['game_id']) & (goalies['team_id'] == row['home_team_id'])].empty:
                homePerc = goalies.loc[(goalies['game_id'] == row['game_id']) & (goalies['team_id'] == row['home_team_id'])]
                homePerc = homePerc['percentage'].values[0]
    if not (gameInput.loc[((gameInput['away_team_id'] == row['away_team_id']) & (gameInput['date_time_GMT'] < row['date_time_GMT']))].tail(10).empty):
        awayLast = gameInput.loc[((gameInput['away_team_id'] == row['away_team_id']) & (gameInput['date_time_GMT'] < row['date_time_GMT']))].tail(10)
        for index1, game in awayLast.iterrows():
            awayS += game['away_goals']
            awayC += game['home_goals']
            if(game['away_goals'] > game['home_goals']):
                awayW += 1
            if not goalies.loc[(goalies['game_id'] == row['game_id']) & (goalies['team_id'] == row['away_team_id'])].empty:
                awayPerc = goalies.loc[(goalies['game_id'] == row['game_id']) & (goalies['team_id'] == row['away_team_id'])]
                awayPerc = awayPerc['percentage'].values[0]
    d = {'game_id': row['game_id'], 'season': row['season'], 'time': row['date_time_GMT'], 'home_scored': homeS, 'home_conceded': homeC, 'home_win': homeW, 'home_percentage': homePerc, 'home_id':row['home_team_id'], 'away_scored': awayS, 'away_conceded': awayC, 'away_win': awayW, 'away_percentage': awayPerc, 'away_id': row['away_team_id']}
    games.at[index, :] = d
#     print(row['game_id'])
        
```

Pri tvorbe tohto datasetu využívame tabuľku game a zároveň aj vytvorený dataFrame s názvom dataFrameGameGoalies. Tento dataFrame sme vytvorili pri tvorbe druhého datasetu.

Vytvoríme si kópiu dataFrame-u, ktorý obsahuje aj percentuálnu úspešnosť brankárov.

```python
games = gameInput1
```

Teraz si musíme vytvoriť dva ďalšie stĺpce, ktoré budú obsahovať, počet víťazstiev domáceho alebo hosťujúceho tímu v predchádzajúcich 5 zápasoch.

```python
games['home_win'] = 0
games['away_win'] = 0
```

```python
games.head()
```

V tomto prípade nejdeme vytvárať nový dataFrame, ale budeme aktualizovať dataFrame games.

Hľadáme posledných 10 zápasov pre domáci aj hosťujúci klub. Vypočítame priemerný počet gólov v týchto zápasoch, a zároveň aj počet výhier z týchto zápasov.

```python
for index, row in games.iterrows():
    homeS = 0;
    homeC = 0;
    homeW = 0;
    awayS = 0;
    awayC = 0;
    awayW = 0;
    if not (gameInput.loc[((gameInput['home_team_id'] == row['home_team_id']) | (gameInput['away_team_id'] == row['home_team_id'])) & (gameInput['date_time'] < row['date_time'])].tail(10).empty):
        homeLast = gameInput.loc[((gameInput['home_team_id'] == row['home_team_id']) | (gameInput['away_team_id'] == row['home_team_id'])) & (gameInput['date_time'] < row['date_time'])].tail(10)
        for index1, game in homeLast.iterrows():
            if(game['home_team_id'] == row['home_team_id']):
                homeS += game['home_goals']
                homeC += game['away_goals']
                if(game['home_goals'] > game['away_goals']):
                    homeW += 1
            elif (game['away_team_id'] == row['home_team_id']):
                homeS += game['away_goals']
                homeC += game['home_goals']
                if(game['away_goals'] > game['home_goals']):
                    homeW += 1
    if not (gameInput.loc[((gameInput['home_team_id'] == row['away_team_id']) | (gameInput['away_team_id'] == row['away_team_id'])) & (gameInput['date_time'] < row['date_time'])].tail(10).empty):
        awayLast = gameInput.loc[((gameInput['home_team_id'] == row['away_team_id']) | (gameInput['away_team_id'] == row['away_team_id'])) & (gameInput['date_time'] < row['date_time'])].tail(10)
        for index1, game in awayLast.iterrows():
            if(game['away_team_id'] == row['away_team_id']):
                awayS += game['away_goals']
                awayC += game['home_goals']
                if(game['away_goals'] > game['home_goals']):
                    awayW += 1
            elif (game['home_team_id'] == row['away_team_id']):
                awayS += game['home_goals']
                awayC += game['away_goals']
                if(game['home_goals'] > game['away_goals']):
                    awayW += 1
    games.loc[index, 'home_scored_avg'] = homeS/10
    games.loc[index, 'away_scored_avg'] = awayS/10
    games.loc[index, 'home_conceded_avg'] = homeC/10
    games.loc[index, 'away_conceded_avg'] = awayC/10
    games.loc[index, 'home_win'] = homeW    
    games.loc[index, 'away_win'] = awayW        
```

Dataset games si rozdelíme do dvoch skupín, na trénovaciu a testovaciu.

```python
dataTrain1 = games.loc[games.season != 20182019]
dataTest1 = games.loc[games.season == 20182019]
```

Následne odstránime nepotrebné stĺpce a spravíme ďalšie rozdelenie, na vstupné vektory a očakávané hodnoty.

```python
dataSet3TrainX = dataTrain1.drop('season',1).drop('game_id',1).drop('date_time',1).drop('total',1)
dataSet3TrainY = dataTrain1.total
dataSet3TestX = dataTest1.drop('season',1).drop('game_id',1).drop('date_time',1).drop('total',1)
dataSet3TestY = dataTest1.total
```

```python
dataSet3TrainX.to_csv('.\datasets\dataSet3TrainX.csv', encoding='utf-8', index=False)
dataSet3TrainY.to_csv('.\datasets\dataSet3TrainY.csv', encoding='utf-8',index=False)
dataSet3TestX.to_csv('.\datasets\dataSet3TestX.csv', encoding='utf-8', index=False)
dataSet3TestY.to_csv('.\datasets\dataSet3TestY.csv', encoding='utf-8', index=False)
```

```python
gameInput['date_time_GMT'] = pd.to_datetime(gameInput['date_time_GMT'], errors='coerce')
```

```python
games = pd.DataFrame(columns=['game_id', 'season', 'time', 'home_scored', 'home_conceded', 'home_win', 'home_id', 'away_scored', 'away_conceded', 'away_win', 'away_id'])
for index, row in gameInput.iterrows():
    homeS = 0;
    homeC = 0;
    homeW = 0;
    awayS = 0;
    awayC = 0;
    awayW = 0;
    if not (gameInput.loc[((gameInput['home_team_id'] == row['home_team_id']) | (gameInput['away_team_id'] == row['home_team_id'])) & (gameInput['date_time_GMT'] < row['date_time_GMT'])].tail(10).empty):
        homeLast = gameInput.loc[((gameInput['home_team_id'] == row['home_team_id']) | (gameInput['away_team_id'] == row['home_team_id'])) & (gameInput['date_time_GMT'] < row['date_time_GMT'])].tail(10)
        for index1, game in homeLast.iterrows():
            if(game['home_team_id'] == row['home_team_id']):
                homeS += game['home_goals']
                homeC += game['away_goals']
                if(game['home_goals'] > game['away_goals']):
                    homeW += 1
            elif (game['away_team_id'] == row['home_team_id']):
                homeS += game['away_goals']
                homeC += game['home_goals']
                if(game['away_goals'] > game['home_goals']):
                    homeW += 1
    if not (gameInput.loc[((gameInput['home_team_id'] == row['away_team_id']) | (gameInput['away_team_id'] == row['away_team_id'])) & (gameInput['date_time_GMT'] < row['date_time_GMT'])].tail(10).empty):
        awayLast = gameInput.loc[((gameInput['home_team_id'] == row['away_team_id']) | (gameInput['away_team_id'] == row['away_team_id'])) & (gameInput['date_time_GMT'] < row['date_time_GMT'])].tail(10)
        for index1, game in awayLast.iterrows():
            if(game['away_team_id'] == row['away_team_id']):
                awayS += game['away_goals']
                awayC += game['home_goals']
                if(game['away_goals'] > game['home_goals']):
                    awayW += 1
            elif (game['home_team_id'] == row['away_team_id']):
                awayS += game['home_goals']
                awayC += game['away_goals']
                if(game['home_goals'] > game['away_goals']):
                    awayW += 1
    
    
    d = {'game_id': row['game_id'], 'season': row['season'], 'time': row['date_time_GMT'], 'home_scored': homeS, 'home_conceded': homeC, 'home_win': homeW, 'home_id':row['home_team_id'], 'away_scored': awayS, 'away_conceded': awayC, 'away_win': awayW, 'away_id': row['away_team_id']}
    games.at[index, :] = d
        
```
